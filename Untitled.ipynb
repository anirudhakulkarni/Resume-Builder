```python
import re
import pandas as pd
import glob
```

```python
import textract
```

```python
text = textract.process("resume.pdf")
text = str(text)

text
```

```python
def proc1(text):
    text = re.split(r"\\n",text)
    for i in range(len(text)):
        text[i] = re.sub(r"\\x0c","",text[i])
        text[i] = re.sub(r'b"',"",text[i])
        text[i] = re.sub(r"b'","",text[i])
        text[i] = re.sub(r"[^\w .-]","",text[i])
    return text
```

```python
listofstudents = []
for f in glob.glob("E:\ResumeBuilding ML\CV-all\CVs Batch of 2017 IITDelhi\*.pdf"):
    try:
        text = textract.process(f)
        listofstudents.append(proc1(str(text)))
    except:
        print("Bad File" , f)
    
        
    
```

```python
len(listofstudents)
```

```python
cols = ["NAME","DEGREE","INSTITUTE","GPAMarks","SCHOLASTIC ACHIEVEMENTS","INTERNSHIPS","PROJECTS","EXTRA CURRICULAR ACTIVITIES","POSITIONS OF RESPONSIBILITY"]
```

```python
cols
```

```python
headers = ['SCHOLASTIC ACHIEVEMENTS','INTERNSHIPS','PROJECTS','EXTRA CURRICULAR ACTIVITIES','POSITIONS OF RESPONSIBILITY']
headers
```

```python
data=[]
for per in listofstudents:
    stud=["" for i in range(len(cols))]
    #NAME
    stud[0] = per[0]
    #DEGREE
    for i in per:
        match=re.search("Tech",i)
        if(match!=None):
            stud[1]= match.string
            break
    #Institute
    for i in per:
        match = re.search(r"\bInstitute(\s|\w)\b",i)
        if(match!=None):
            stud[2] = match.string
            break
    #GPAMarks
    for i in range(len(per)):
        k=per[i].lower()
        match=re.search("gpamarks|marks",k)
        if(match!=None):
            j=i+1
            while(j<i+10):
                try:
                    k = float(per[j])
                    if(k>0 and k<10):
                        stud[3]=k
                        break
                    else:
                        j+=1
                except:
                    j+=1
    
    #SCHOLASTIC ACHIEVEMENTS
    for i in range(len(per)):
        k = per[i]
        match=re.search(headers[0],k)
        if(match):
            leng= []
            j = i+1
            condition = "|".join(i for i in headers)
            while(j<len(per) and not re.search(condition,per[j])):
                leng.append(per[j])
                j+=1
            stud[4]=leng
            break
    #INTERNSHIPS
    for i in range(len(per)):
        k = per[i]
        match=re.search(headers[1],k)
        if(match):
            leng= []
            j = i+1
            condition = "|".join(i for i in headers)
            while(j<len(per) and not re.search(condition,per[j])):
                leng.append(per[j])
                j+=1
            stud[5]=leng
            break
    #PROJECTS
    for i in range(len(per)):
        k = per[i]
        match=re.search(headers[2],k)
        if(match):
            leng= []
            j = i+1
            condition = "|".join(i for i in headers)
            while(j<len(per) and not re.search(condition,per[j])):
                leng.append(per[j])
                j+=1
            stud[6]=leng
            break
    #EXTRA CURRICULAR ACTIVITIES
    for i in range(len(per)):
        k = per[i]
        match=re.search(headers[3],k)
        if(match):
            leng= []
            j = i+1
            condition = "|".join(i for i in headers)
            while(j<len(per) and not re.search(condition,per[j])):
                leng.append(per[j])
                j+=1
            stud[7]=leng
            break
    #POSITION OF RESPOSIBILITIES
    for i in range(len(per)):
        k = per[i]
        match=re.search(headers[4],k)
        if(match):
            leng= []
            j = i+1
            condition = "|".join(i for i in headers)
            while(j<len(per) and not re.search(condition,per[j])):
                leng.append(per[j])
                j+=1
            stud[8]=leng
            break
    data.append(stud)

    
```
